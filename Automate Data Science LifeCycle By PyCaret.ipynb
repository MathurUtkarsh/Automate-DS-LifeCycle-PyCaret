```python
# load dataset
from pycaret.datasets import get_data 
boston = get_data('boston') 
```

```python
# init setup
from pycaret.regression import * 
reg1 = setup(data = boston, target = 'medv')
```

```python
best = compare_models()
```

```python
print(best)
```

```python
evaluate_model(best)
```

```python
# train model
gbr = create_model('gbr')
```

```python
# tune model
tuned_gbr = tune_model(gbr)
```
